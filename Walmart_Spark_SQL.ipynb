```python
#Q1. Number of stores in each store Type

spark.sql("Select Type, count(store) store_count from store group by Type order by store_count  desc").show()
```

```python
#Q2.Total sale by store type

spark.sql("select Type, count('Avg Yearly Sales (Million)') as sales from store group by Type order by sales desc").show()
```

```python
#Q3.Top 10 stores which are making the highest Sales.

spark.sql(“select Store, count(“Avg Yearly Sales (Million)”) as Store_sales from stores group by Store order by Store_sales desc limit 10”).show()
```

```python
#Q4.Bottom 10 stores which are making the highest Sales.

spark.sql(“select Store, count(“Avg Yearly Sales (Million)”) as salesByStores from stores group by Store order by salesByStores limit 10”).show()
```

```python
#Q5. What is the max and min weekly sales?
spark.sql(“select(max('Avg Yearly Sales (Million)'), min('Weekly Sales (Million)') from stores").show()
```

```python
#Q6. Total sales on week of Holiday And week of not having Holiday.

spark.sql(“select IsHoliday, count(“Avg Yearly Sales (Million)”) as Sales from store_sales group by IsHoliday order by Sales limit 10”).show()
```

```python
#Q7.Top 10 stores by Markdown.

spark.sql(“select Store, count(“MarkDown”) as Total_MarkDown from store_sales
group by Store order by Total_MarkDown limit 10”).show()
```

```python
#Q8.Find how much Sales is done per Year
spark.sql(“select year, count(“Store Weekly Sales (Thousand)”) as Sales_per_Year from store_sales
group by year order by Sales_per_Year”).show()
```
